# Defining the Model

```python
def Convolution(input_tensor, filters):
    # Apply a 2D convolution layer with specified number of filters, 3x3 kernel, same padding,
    # and a stride of (1,1). Also, apply L2 regularization with a factor of 0.001.
    x = Conv2D(filters=filters, kernel_size=(3, 3), padding="same", strides=(1, 1), kernel_regularizer=l2(0.001))(input_tensor)
    
    # Apply a dropout layer with a dropout rate of 0.1 to the output of the convolution.
    x = Dropout(0.1)(x)
    
    # Apply a ReLU activation function to the output of the dropout layer.
    x = Activation('relu')(x)
    
    # Return the final output tensor.
    return x
```

```python
def model(input_shape):                    # Define the input layer with the specified input shape
    
    inputs = Input((input_shape))
    
    # Apply the first convolutional block with 32 filters and a 2x2 max pooling layer
    conv_1 = Convolution(inputs, 32)
    maxp_1 = MaxPooling2D(pool_size=(2, 2))(conv_1)
    
    # Apply the second convolutional block with 64 filters and a 2x2 max pooling layer
    conv_2 = Convolution(maxp_1, 64)
    maxp_2 = MaxPooling2D(pool_size=(2, 2))(conv_2)
    
    # Apply the third convolutional block with 128 filters and a 2x2 max pooling layer
    conv_3 = Convolution(maxp_2, 128)
    maxp_3 = MaxPooling2D(pool_size=(2, 2))(conv_3)
    
    # Apply the fourth convolutional block with 256 filters and a 2x2 max pooling layer
    conv_4 = Convolution(maxp_3, 256)
    maxp_4 = MaxPooling2D(pool_size=(2, 2))(conv_4)
    
    # Flatten the output from the last max pooling layer
    flatten = Flatten()(maxp_4)
    
    # Apply two dense layers with 64 units and ReLU activation
    dense_1 = Dense(64, activation='relu')(flatten)
    dense_2 = Dense(64, activation='relu')(flatten)
    
    # Apply a dropout layer with a dropout rate of 0.2 to each dense layer
    drop_1 = Dropout(0.2)(dense_1)
    drop_2 = Dropout(0.2)(dense_2)
    
    # Define two output layers:
    output_1 = Dense(1, activation='sigmoid', name='sex_out')(drop_1)    # The first output layer is a dense layer with 1 unit and sigmoid activation, named 'sex_out'
    output_2 = Dense(1, activation='relu', name='age_out')(drop_2)      # The second output layer is a dense layer with 1 unit and ReLU activation, named 'age_out'
    
    model = Model(inputs=[inputs], outputs=[output_1, output_2])   # Create the model with the specified inputs and outputs
    
    # Compile the model with binary cross-entropy loss for 'sex_out' and mean absolute error loss for 'age_out',
    model.compile(loss=["binary_crossentropy", "mae"], optimizer="Adam", metrics=["accuracy","accuracy"])               # using the Adam optimizer and accuracy as a metric
    return model # Return the compiled model
```

```python
# Instantiate the model with an input shape of (48, 48, 3) using the defined model function
Model = model((48, 48, 3))
```

```python
# Print a summary of the model, displaying the layers and their shapes
Model.summary()
```

```python
# Import the ModelCheckpoint callback from TensorFlow Keras
from tensorflow.keras.callbacks import ModelCheckpoint # type: ignore
```

# Initializing the Model

```python
fle_s='Age_Gender_Detection.keras'
checkpoint=ModelCheckpoint(fle_s,monitor='val_loss',verbose=1,save_best_only=True,save_weights_only=False,mode='auto',save_freq='epoch')
Early_stop=tf.keras.callbacks.EarlyStopping(patience=75,monitor='val_loss',restore_best_weights='True')
callback_list=[checkpoint,Early_stop]
```

```python
# Train the model on the training data (X_train, Y_train_2) with a batch size of 64,
# validating on the test data (X_test, Y_test_2), for 250 epochs, and using the defined callbacks
History = Model.fit(X_train, Y_train_2, batch_size=64, validation_data=(X_test, Y_test_2),epochs=250, callbacks=callback_list)
```
